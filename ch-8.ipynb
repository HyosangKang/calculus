# 8장. 선적분과 면적분 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 8.1 벡터장과 선적분 

```python
# 벡터장을 화살표로 그리기 
intv = np.linspace(-1, 1, 6)
pt = np.array(np.meshgrid(intv, intv)).reshape(2, -1)
f = lambda x, y: np.array([x,x*y**2])
for i in range(pt.shape[1]):
    plt.quiver(*pt[:,i], *f(*pt[:,i]))
plt.xlim(-1.25, 1.25)
plt.ylim(-1.25, 1.25)
```

```python
# 벡터장을 화살표로 그리기 
intv = np.linspace(-1, 1, 6)
pt = np.array(np.meshgrid(intv, intv)).reshape(2, -1)
f = lambda x, y: np.array([x*y**2,-x*y])
for i in range(pt.shape[1]):
    plt.quiver(*pt[:,i], *f(*pt[:,i]))
plt.xlim(-1.25, 1.25)
plt.ylim(-1.25, 1.25)
```
