```python
t = np.linspace(0, np.pi/4)
x = lambda t: np.cos(np.cos(t)) + np.sin(t)
y = lambda t: np.sin(np.sin(t)+1)
plt.plot(x(t), y(t))
t0, t1 = .2, np.pi/4 - .2
plt.plot([x(t0), x(t1)], [y(t0), y(t1)], 'ko-')
plt.text(x(t0), y(t0)+.05, r'$c_{i-1} = c(t_{i-1|})$', fontsize=15, usetex=True, ha='center')
plt.text(x(t1), y(t1)+.05, r'$c_i = c(t_i)$', fontsize=15, usetex=True, ha='center')
plt.text(x(t0)+.15, y(t1)-.1, r'$\Vert c_{i-1}c_i\Vert$', fontsize=15, usetex=True)
plt.text(x(np.pi/4), y(np.pi/4)+.03, r'$c(t)$', fontsize=15, usetex=True, ha='right')
t2 = np.pi/4
plt.plot([x(t1), x(t1)+1*(x(t2)-x(t1))], [y(t1), y(t1)+1*(y(t2)-y(t1))-.01], 'k')
t2 = 0
plt.plot([x(t0), x(t0)+1*(x(t2)-x(t0))], [y(t0), y(t0)+1*(y(t2)-y(t0))-.01], 'k')
plt.plot(x((t0+t1)/2), y((t0+t1)/2), 'ko')
plt.text(x((t0+t1)/2), y((t0+t1)/2)+.05, r'$c(t_i^*)$', fontsize=15, usetex=True, ha='center')  

plt.axis('off')
plt.axis('equal')
plt.savefig('img/ch-8/8-1-1-1.png', dpi=600, bbox_inches='tight')
```

```python
t = np.linspace(0, 2*np.pi)
th = np.pi/4
pt = np.array([np.cos(th)*t - np.sin(th)*np.sin(t), np.sin(th)*t + np.cos(th)*np.sin(t)])
ax = plt.figure().add_subplot(projection='3d')
ax.plot(pt[0], pt[1], np.zeros_like(t))

f = lambda x, y: np.sin(x) +2
# draw surface bounded by the curve and the graph of f
x = np.array([pt[0], pt[0]])
y = np.array([pt[1], pt[1]])
z = np.array([np.zeros_like(t), f(pt[0], pt[1])])
ax.plot_surface(x, y, z, alpha=.5, edgecolor='k', lw=.3, color='C0')
ax.plot(pt[0], pt[1], f(pt[0], pt[1]), 'k')
ax.text(pt[0, 20], pt[1, 20], f(pt[0, 20], pt[1, 20])+.3, r'$z = f(c(t))$', fontsize=15, usetex=True)
ax.text(pt[0, 20], pt[1, 20], 0-.56, r'$c(t)$', fontsize=15, usetex=True)

plt.savefig('img/ch-8/8-1-1-2.png', dpi=600, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(10, 5)
intv = np.linspace(0, 1, 6)
uu, vv = np.meshgrid(np.linspace(0, 1, 6), np.linspace(0, 1, 6))
for i in range(6):
    ax[0].plot([0, 1], [intv[i], intv[i]], 'k--')
    ax[0].plot([intv[i], intv[i]], [0, 1], 'k--')
ax[0].fill([0, 1, 1, 0], [0, 0, 1, 1], 'C0', alpha=.3)
ax[0].set_title('UV-plane')
ax[0].set_xlim([-.25, 1.25])
ax[0].set_ylim([-.25, 1.25])
ax[0].axis('off')

s = lambda u, v: np.array([u, v, u**2+v**2])
ax[1].axis('off')
ax[1] = fig.add_subplot(122, projection='3d')
ax[1].set_title('XYZ-space')
for i in range(6):
    t = np.linspace(0, 1)
    ax[1].plot(*s(t, intv[i]*np.ones_like(t)), 'k--')
    ax[1].plot(*s(intv[i]*np.ones_like(t), t), 'k--')

su = lambda u, v: np.array([1, 0, 2*u])
sv = lambda u, v: np.array([0, 1, 2*v])
for i in range(5):
    for j in range(5):
        u, v = uu[i,j], vv[i,j]
        p = s(u,v)
        a, b = su(u,v)/6, sv(u,v)/6
        tt, ss = np.meshgrid(np.linspace(0, 1), np.linspace(0, 1))
        pt = p[:, np.newaxis, np.newaxis] + a[:, np.newaxis, np.newaxis]*tt + b[:, np.newaxis, np.newaxis]*ss
        ax[1].plot_surface(*pt, alpha=.3, color='C0')

plt.savefig('img/ch-8/8-2-1-1.png', dpi=600, bbox_inches='tight')
```

```python
import pygame
import numpy as np

Width, Height = 800, 600

def init_goal():
    min_vert, max_vert = 4, 10
    min_area, max_area = 500, 4500
    target_vertices = np.random.randint(min_vert, max_vert)
    target_area = int(min_area + (max_area - min_area)*np.random.rand())
    return target_vertices, target_area

def display_goal(screen, target_vertices, target_area):
    text = font.render(f'Number of Vertex: %d'%target_vertices, True, (255, 255, 255))
    screen.blit(text, (Width-350 , Height-100))
    text = font.render(f'Target Area: %d'%target_area, True, (255, 255, 255))
    screen.blit(text, (Width-350 , Height-50))

def display_area(screen, xy):
    text = font.render(f'%.1f'%area(xy), True, (255, 255, 255))
    screen.blit(text, (20, 5))

def area(xy):
    xy = np.array(xy)/10
    xy -= np.array([40, 30])
    sum = 0
    for i in range(len(xy)):
        sum += xy[i][0]*xy[(i+1)%len(xy)][1] - xy[i][1]*xy[(i+1)%len(xy)][0]
    return abs(sum)/2

pygame.init()
pygame.font.init()
font = pygame.font.SysFont('Comic Sans MS', 30)
screen = pygame.display.set_mode((Width, Height))

play, num_vertices, xy = True, 0, []
target_vertices, target_area = init_goal()
display_goal(screen, target_vertices, target_area)

while play:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            play = False
        if event.type == pygame.MOUSEBUTTONDOWN:
            if pygame.mouse.get_pressed()[0]:
                if num_vertices < target_vertices:
                    num_vertices += 1
                    x, y = pygame.mouse.get_pos()
                    xy.append((x, y))
                    pygame.draw.circle(screen, (255, 255, 255), (x, y), 3)
                    if num_vertices == target_vertices:
                        pygame.draw.polygon(screen, (255, 255, 255), xy)
                        display_area(screen, xy)
                else:
                    screen.fill((0, 0, 0))
                    num_vertices, xy = 0, []
                    target_vertices, target_area = init_goal()
                    display_goal(screen, target_vertices, target_area)
    pygame.display.flip()

pygame.quit()
```
