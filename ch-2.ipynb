```python
v1 = np.array([1, 1, 2])
v2 = np.array([0, 1, 3])
w = np.array([1, 0, 2])
theta_1 = np.arccos(np.dot(v1, w) / np.sqrt(np.dot(w, w)*np.dot(v1, v1)))
theta_2 = np.arccos(np.dot(v2, w) / np.sqrt(np.dot(w, w)*np.dot(v2, v2)))

ax = plt.figure().add_subplot(projection='3d')
t = np.linspace(0, 2*np.pi)
r = np.linspace(-1/np.tan(theta_1), 1/np.tan(theta_1))
T, R = np.meshgrid(t, r)
e1 = v1 / np.sqrt(np.dot(v1, v1))
e2 = np.cross(v1, w)
e2 = e2 / np.sqrt(np.dot(e2, e2))
e3 = np.cross(e1, e2)
e3 = e3 / np.sqrt(np.dot(e3, e3))
xyz_1 = R[np.newaxis, :, :] * e1[:, np.newaxis, np.newaxis] 
xyz_2 = np.tan(theta_1) * R[np.newaxis, :, :] 
xyz_3 = np.cos(T)[np.newaxis, :, :] * e2[:, np.newaxis, np.newaxis] + np.sin(T)[np.newaxis, :, :] * e3[:, np.newaxis, np.newaxis]
xyz = xyz_1 + xyz_2 * xyz_3
ax.plot_surface(*xyz, color='gray', alpha=0.3, edgecolor='k', lw=.3)
ax.quiver(0, 0, 0, *v1, color='k', arrow_length_ratio=.1, lw=3)
ax.text(*(v1 + np.array([0, 0, .3])), r'$v_1$', **text_params_latex)
t = np.linspace(0, 2*np.pi)
r = np.linspace(-1/np.tan(theta_2), 1/np.tan(theta_2))
T, R = np.meshgrid(t, r)
e1 = v2 / np.sqrt(np.dot(v2, v2))
e2 = np.cross(v2, w)
e2 = e2 / np.sqrt(np.dot(e2, e2))
e3 = np.cross(e1, e2)
e3 = e3 / np.sqrt(np.dot(e3, e3))
xyz_1 = R[np.newaxis, :, :] * e1[:, np.newaxis, np.newaxis] 
xyz_2 = np.tan(theta_2) * R[np.newaxis, :, :] 
xyz_3 = np.cos(T)[np.newaxis, :, :] * e2[:, np.newaxis, np.newaxis] + np.sin(T)[np.newaxis, :, :] * e3[:, np.newaxis, np.newaxis]
xyz = xyz_1 + xyz_2 * xyz_3
ax.plot_surface(*xyz, color='gray', alpha=0.3, edgecolor='k', lw=.3)
ax.quiver(0, 0, 0, *v2, color='k', arrow_length_ratio=.1, lw=3)
ax.text(*(v2 + np.array([0, 0, .3])), r'$v_2$', **text_params_latex)

ax.quiver(0, 0, 0, 1, 0, 2, color='C0', arrow_length_ratio=.1, lw=3)
ax.quiver(0,0,0 , 5/11, 18/11, 16/11, color='C0', arrow_length_ratio=.1, lw=3)
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
ax.set_box_aspect([1,1,1])
# ax.set_aspect('equal')
ax.view_init(elev=20., azim=52)

plt.savefig('img/ch-2/extra-7.png', dpi=1200, bbox_inches='tight')
```

```python
def det(a: np.array) -> float:
    if len(a.shape) != 2 or a.shape[0] != a.shape[1]:
        raise ValueError('The input matrix must be square')
    if a.shape[0] == 1:
        return a[0, 0]
    d = 0
    for j in range(a.shape[1]):
        sg = 1 if j % 2 == 0 else -1
        jdx = [k for k in range(a.shape[1]) if k != j]
        d += sg * a[0, j] * det(a[1:][:, jdx])
    return d
```

```python
def inv(a: np.array) -> np.array:
    if len(a.shape) != 2 or a.shape[0] != a.shape[1]:
        raise ValueError('The input matrix must be square')
    d = det(a)
    if d == 0:
        raise ValueError('The input matrix is singular')
    invA = np.empty(a.shape)
    for i in range(a.shape[0]):
        for j in range(a.shape[1]):
            sg = 1 if (i + j) % 2 == 0 else -1
            idx = [k for k in range(a.shape[0]) if k != i]
            jdx = [k for k in range(a.shape[1]) if k != j]
            invA[j, i] = sg * det(a[idx][:, jdx]) / d
    return invA
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def chr2im(s) -> np.ndarray:
    plt.figure(figsize=(1, 1))
    plt.text(0, 0, s, fontsize=80, ha='center', va='center')
    plt.axis('off')
    plt.axis('tight')
    plt.savefig('temp.png', bbox_inches='tight')
    img = plt.imread('temp.png')
    plt.clf()
    plt.close()
    os.remove('temp.png')
    return img
```

```python
def im2arr(im) -> np.ndarray:
    h, w, l = im.shape
    m = np.array([[k for k in range(w) for _ in range(h)], 
                  [h - 1 - k for _ in range(w) for k in range(h)]])
    m = np.diag([1/w, 1/h]) @ m - np.array([0.5, 0.5])[:, np.newaxis]
    im = im.reshape(h*w, l, order='F')
    black = [i for i, val in enumerate(im) if np.prod(val) < 1.0 ]
    return m[:, black]
```
