## 2.1 벡터 

```python
# pyplot의 quiver 함수를 이용해 벡터를 표현
# 각도 지정 방법(angles), 크기 지정 방법(scale_units), 크기(scale), 두께(width)를 설정
plt.quiver(1, 1, 2, 3, angles='xy', scale_units='xy', scale=1, width=.01)
plt.xlim(0, 4)                 # x축 범위를 [0, 4]로 설정
plt.ylim(0, 5)                 # y축 범위를 [0, 5]로 설정
plt.gca().set_aspect('equal')  # x축과 y축의 비율을 동일하게 설정
```

```python
# 3차원 벡터를 이미지로 표현
ax = plt.subplot(projection='3d') # 3차원 이미지를 표현하도록 설정
# quiver 메소드(method)를 이용해 3차원 벡터를 표현
# 화살촉 크기비(arrow_length_ratio), 두께(lw), 색(color)을 설정
ax.quiver(1, 1, 2, 2, 3, 4, arrow_length_ratio=0.1, lw=3, color='k')
ax.set_xlim([0, 4])    # x축 범위를 [0, 4]로 설정
ax.set_ylim([0, 5])    # y축 범위를 [0, 5]로 설정
ax.set_zlim([0, 6])    # z축 범위를 [0, 6]으로 설정
ax.set_aspect('equal') # 각 축의 축척을 같게 설정
```

```python
# 2개 이상의 2차원 벡터를 이미지로 표현
# quiver 함수의 옵션값을 딕셔너리값(dict)으로 저장
quiver_2d_params = dict(angles='xy', scale_units='xy', scale=1, width=.01)

# quiver 함수에 두 벡터의 시작점의 x, y좌표, 각 축 방향으로의 크기를 입력
plt.quiver([1, 0], [1, 0], [1, 1], [2, 2], **quiver_2d_params)

# 두 벡터의 시작점과 끝점을 검은색 동그라미(ko)로 표시
plt.plot([1, 0, 2, 1], [1, 0, 3, 2], 'ko')

# text 함수를 이용해 각 점 주변에 점의 이름을 문자로 출력
plt.text(1.1, 1, 'P', fontsize=15, va='center', ha='left')
plt.text(0.1, 0, 'O', fontsize=15, va='center', ha='left')
plt.text(2.1, 3, 'Q', fontsize=15, va='center', ha='left')
plt.text(1.1, 2, 'R', fontsize=15, va='center', ha='left')
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.gca().set_aspect('equal')
```

```python
# ndarray 객체를 이용해 위치벡터를 표현
p = np.array([1, 1])   # 위치벡터의 시작점을 저장
pq = np.array([2, 3])  # 위치벡터의 좌푯값을 ndarray 값으로 저장
plt.quiver(p[0], p[1], pq[0], pq[1], **quiver_2d_params)
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.gca().set_aspect('equal')
```

```python
# ndarray 객체의 포인터를 이용해 위치벡터를 표현
p = [1, 1]
pq = np.array([2, 3])
plt.quiver(*p, *pq, **quiver_2d_params)  # 값 대신 포인터를 입력
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.gca().set_aspect('equal')
```

```python
# ndarray 객체를 이용해 2개 이상의 위치벡터를 이미지로 표현
xy = [[1, 0], [1, 0]]   # 두 위치벡터의 시작점 (1, 1), (0, 0)을 저장
uv = [[1, 1], [2, 2]]   # 두 위치벡터의 좌푯값 (1, 2), (1, 2)를 저장
plt.quiver(*xy, *uv, **quiver_2d_params) # 포인터를 참조하여 위치벡터를 그림
plt.xlim(-1, 3)
plt.ylim(-1, 4)
```

```python
# 5개의 2차원 위치벡터와 이름을 이미지로 표현
quiver_start_xy = [[0, 1, 0, 1, 0], [0, 1, 0, 2, 0]]
quiver_xy = np.array([[1, 1, 1, 1, 2], [1, 2, 2, 1, 3]])
colors = ['k', 'k', 'k', 'k', 'C0']
plt.quiver(*quiver_start_xy, *quiver_xy, color=colors, **quiver_2d_params)

point_xy = [[0, 1, 2, 1], [0, 1, 3, 2]]  # 레이블을 표시할 점의 위치를 저장
label = ['O', 'P', 'Q', 'R']             # 레이블의 값(문자)을 순서대로 저장
plt.plot(*point_xy, 'ko')                # plot 함수에 포인터 정보를 전달하여 점을 표시

# text 함수의 옵션을 딕셔너리 값으로 저장
text_params = dict(fontsize=15, va='center', ha='left')

# 각 점의 순서대로 x축 값에서 0.1만큼 떨어진 곳에 레이블을 표시
for i in range(4):
    plt.text(point_xy[0][i]+.1, point_xy[1][i], label[i], **text_params)

plt.gca().set_aspect('equal')
```

```python
# 행렬식을 계산하는 함수를 구현
def det(a: np.array) -> float:
    # 2차원 배열이 아니거나 정방행렬이 아니면 에러를 출력
    if len(a.shape) != 2 or a.shape[0] != a.shape[1]:
        raise ValueError('The input matrix must be square')
    # 1x1 행렬인 경우 성분을 그대로 반환
    if a.shape[0] == 1:
        return a[0, 0]
    # 0행을 기준으로 행렬식을 계산
    d = 0
    # 0행의 j열의 성분마다 계산을 반복 수행
    for j in range(a.shape[1]):
        # j가 짝수이면 부호(sg)를 1로, 홀수이면 -1로 설정
        sg = 1 if j % 2 == 0 else -1
        # j열을 제외한 모든 열 인덱스를 저장
        jdx = [k for k in range(a.shape[1]) if k != j]
        # 부호, 0행의 j열 성분, 부분행렬의 행렬식을 곱하여 더함
        d += sg * a[0, j] * det(a[1:][:, jdx])
    return d
```

```python
# 역행렬을 계산하는 함수를 구현
def inv(a: np.array) -> np.array:
    if len(a.shape) != 2 or a.shape[0] != a.shape[1]:
        raise ValueError('The input matrix must be square')
    # 행렬식이 0이면 오류를 출력
    d = det(a)
    if d == 0:
        raise ValueError('The input matrix is singular')
    invA = np.empty(a.shape)
    for i in range(a.shape[0]):
        for j in range(a.shape[1]):
            sg = 1 if (i + j) % 2 == 0 else -1
            idx = [k for k in range(a.shape[0]) if k != i]
            jdx = [k for k in range(a.shape[1]) if k != j]
            invA[j, i] = sg * det(a[idx][:, jdx]) / d
    return invA
```

```python
# 역행렬을 이용해 연립방정식의 해 구하기
b = np.array([[1 , 2, 1], [-1, 0, 1], [0, 1, 2]])
print(inv(b)@np.array([3, 1, -1]))
```

## 2.4 프로젝트: 캡차(CAPTCHA) 만들기 

```python
# 하나의 문자를 픽셀 이미지 정보(ndarray)로 변환하는 함수를 구현
import os

def chr2im(s) -> np.ndarray:
    plt.figure(figsize=(1, 1))
    # 문자를 윈도우에 출력
    plt.text(0, 0, s, fontsize=80, ha='center', va='center')
    plt.axis('off')    # 좌표축을 제거
    plt.axis('tight')  # 좌표축과의 여백을 제거
    plt.savefig('temp.png', bbox_inches='tight')  # 이미지를 임시파일에 저장
    img = plt.imread('temp.png')                  # 이미지를 픽셀 정보로 읽기
    plt.clf()    # 다음 이미지 출력을 위해 이미지 윈도우의 모든 내용을 삭제
    plt.close()  # 이미지 윈도우 닫기
    os.remove('temp.png')  # 임시파일을 삭제
    return img
```

```python
# 숫자 7을 이미지로 출력
img = chr2im('7')
plt.imshow(img)
```

```python
# 픽셀 이미지에서 흰색을 제외한 부분만 추출하는 함수를 구현
def im2arr(im) -> np.ndarray:
    h, w, l = im.shape
    m = np.array([[k for k in range(w) for _ in range(h)], 
                  [h - 1 - k for _ in range(w) for k in range(h)]])
    # 가로, 세로 길이를 1로 맞추고 중심이 원점이 되도록 평행이동
    m = np.diag([1/w, 1/h]) @ m - np.array([0.5, 0.5])[:, np.newaxis]
    im = im.reshape(h*w, l, order='F')
    # 흰색이 아닌 픽셀의 인덱스만 추출
    black = [i for i, val in enumerate(im) if np.prod(val) < 1.0 ]
    return m[:, black]
```

```python
# 대각행렬을 이용해 이미지 변환
im = chr2im('7')
xy = im2arr(im)
fig, ax = plt.subplots(1, 3)
for i in range(3):
    ax[i].set_aspect('equal')
ax[0].scatter(*xy, s=40, color='k')
xy = np.diag([.8, 1.2]) @ xy
ax[1].scatter(*xy, s=40, color='C0')
xy = r(15/180*np.pi) @ xy
ax[2].scatter(*xy, s=40, color='C0')
plt.tight_layout() # 이미지 간격 자동 조절
```

```python
# 숫자로 이루어진 캡차 이미지를 출력
import random 

plt.gca().set_aspect('equal')
plt.axis('tight')
plt.axis('off')

text = '1234'  # 출력할 문자열을 지정
off = .4       # 각 문자 사이의 간격을 지정
color = np.array([31/255, 119/255, 180/255])
for i, c in enumerate(text):
    im = chr2im(c)   # 문자를 이미지로 변환
    xy = im2arr(im)  # 이미지 픽셀의 좌표 추출
    ld = .8 + .4 * np.random.rand(2)  # 대각성분 값을 무작위로 설정
    xy = np.diag(ld) @ xy             # 대각행렬로 이미지를 변환
    th = -np.pi/4 + np.pi/2 * random.random()  # 회전각을 무작위로 설정
    xy = r(th) @ xy                            # 회전변환행렬로 이미지를 변환
    plt.scatter(*xy + np.array([off*i, 0])[:, None], s=10, color=color)
    
plt.xlim(-off, len(text)*off)
plt.ylim(-.5, .5)
```
